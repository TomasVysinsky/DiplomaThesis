```python
import torch
import torchvision.models as models
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import torch.nn.functional as F

from Occlusion import Occlusion

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

weights = models.ResNet50_Weights.DEFAULT
model = models.resnet50(weights=weights).to(device)
model.eval()

categories = weights.meta["categories"]
```

```python
print("CUDA available:", torch.cuda.is_available())
print("CUDA version:", torch.version.cuda)
print("GPU count:", torch.cuda.device_count())
print("GPU name:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU")
```

```python
explainer = Occlusion(model, device=device)

# Occlusion parameters (tradeoff: smaller stride -> better resolution but slower)
patch_size = 32
stride = 16
```

```python
preprocess = weights.transforms()

def load_image(path):
    img = Image.open(path).convert("RGB")
    tensor = preprocess(img).unsqueeze(0)
    return img, tensor
```

```python
img, input_tensor = load_image("../Pictures/Indian_Cobra.jpg")
input_tensor = input_tensor.to(device)

# Single source of truth for target + confidence
with torch.no_grad():
    logits = model(input_tensor)
    probs = torch.softmax(logits, dim=1)

target_idx = int(torch.argmax(probs, dim=1).item())
confidence = float(probs[0, target_idx].item())
label = categories[target_idx]

occ_map, _, drops_grid = explainer.explain(
    input_tensor,
    target_class=target_idx,
    patch_size=patch_size,
    stride=stride,
    occlusion_value=0.0,
    mode="prob_drop",
    batch_size=32,
)

print(f"Predicted class: {label}")
print(f"Confidence: {confidence:.4f}")
```

```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt

def denormalize_for_display(t, weights):
    mean = torch.tensor(weights.transforms().mean).view(1,3,1,1).to(t.device)
    std = torch.tensor(weights.transforms().std).view(1,3,1,1).to(t.device)
    x = (t * std) + mean
    x = x.clamp(0, 1)
    x = (x[0].permute(1,2,0).detach().cpu().numpy() * 255).astype(np.uint8)
    return x

def overlay(img_uint8, hm, alpha=0.45):
    hm_rgba = plt.get_cmap("jet")(hm)
    hm_rgb = (hm_rgba[..., :3] * 255).astype(np.uint8)
    return (img_uint8 * (1 - alpha) + hm_rgb * alpha).astype(np.uint8)

img_disp = denormalize_for_display(input_tensor, weights)

# comparable map (224x224 aligned to model input)
occ_hm = occ_map[0,0].detach().cpu().numpy()

# comparable overlay
occ_overlay_raw = overlay(img_disp, occ_hm, alpha=0.45)

# legacy enhanced overlay on the same new standardized display
occ_hm_legacy = np.power(occ_hm, 0.7)
occ_overlay_legacy = overlay(img_disp, occ_hm_legacy, alpha=0.5)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img_disp)
plt.title("Original (model input view)")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(occ_overlay_raw)
plt.title(f"Occlusion - comparable\n{label} ({confidence:.2%})")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(occ_overlay_legacy)
plt.title(f"Occlusion - enhanced legacy\npatch={patch_size}, stride={stride}")
plt.axis("off")

plt.tight_layout()
plt.show()
```
