```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt

from Occlusion import Occlusion

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
model = model.to(device)
model.eval()
```

```python
explainer = Occlusion(model, device=device)

# Occlusion parameters (tradeoff: smaller stride -> better resolution but slower)
patch_size = 32
stride = 16
```

```python
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

def load_image(path):
    img = Image.open(path).convert("RGB")
    tensor = transform(img).unsqueeze(0)
    return img, tensor
```

```python
img, input_tensor = load_image("../Pictures/Indian_Cobra.jpg")
input_tensor = input_tensor.to(device)

# Explain (target_class defaults to argmax on original)
occ_map, output, drops_grid = explainer.explain(
    input_tensor,
    target_class=None,
    patch_size=patch_size,
    stride=stride,
    occlusion_value=0.0,
    mode="prob_drop",
    batch_size=32,
)

# Softmax on original output for label + confidence
probs = torch.softmax(output, dim=1)
confidence, pred_class = torch.max(probs, dim=1)

confidence = confidence.item()
pred_class = pred_class.item()

# ImageNet labels
weights = models.ResNet50_Weights.DEFAULT
class_names = weights.meta["categories"]
label = class_names[pred_class]

print(f"Predicted class: {label}")
print(f"Confidence: {confidence:.4f}")
```

```python
import torch.nn.functional as F
import numpy as np

img_np = np.array(img)

# occ_map is (1,1,224,224) aligned to input; resize to original image display size
occ_resized = F.interpolate(
    occ_map,
    size=(img_np.shape[0], img_np.shape[1]),
    mode="bilinear",
    align_corners=False
).squeeze().detach().cpu().numpy()

# Optional: make the heatmap pop a bit (gamma < 1 boosts mid-values)
occ_resized = np.power(occ_resized, 0.7)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img_np)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_np)
plt.imshow(occ_resized, cmap='jet', alpha=0.5)
plt.title(f"Occlusion: {label} ({confidence:.2%})\npatch={patch_size}, stride={stride}")
plt.axis("off")

plt.show()
```
